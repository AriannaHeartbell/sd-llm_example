```python
from transformers import AutoModelForCausalLM, CodeGenTokenizerFast as Tokenizer
from PIL import Image

model_id = "vikhyatk/moondream1"
model = AutoModelForCausalLM.from_pretrained(model_id, trust_remote_code=True)
tokenizer = Tokenizer.from_pretrained(model_id)
```

```python
IMAGE_PATH = '/content/yuina_sg (5)-0000 (1).png'
QUESTION = "1) what is sex of the character? 2)what clothes character wear? 3) what color the character hair is?"
image = Image.open(IMAGE_PATH)
enc_image = model.encode_image(image)
print(model.answer_question(enc_image, QUESTION, tokenizer))
```

```python
IMAGE_PATH = '/content/yuina_sg (5)-0000 (1).png'
QUESTION = "describe the image with detail as you can"
image = Image.open(IMAGE_PATH)
enc_image = model.encode_image(image)
print(model.answer_question(enc_image, QUESTION, tokenizer))
```
